## Data cleaning

### Subtask:
Clean the World Bank Indicators dataset by handling missing values and addressing inconsistencies.


**Reasoning**:
Clean the data by handling missing values, correcting data types, removing duplicates, and optionally addressing outliers.



```python
# Missing Value Imputation
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    elif df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)

# Data Type Handling
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            if df[col].isnull().any():
              df[col].fillna(df[col].median(), inplace=True)
        except (ValueError, TypeError):
            pass  # Skip if conversion fails

# Duplicate Removal
df.drop_duplicates(inplace=True)

# Outlier Detection and Handling (Optional - Example using IQR for one column)
# This is an example, and you may want to apply different outlier handling methods
# to other columns based on their characteristics.
Q1 = df['GDP (current US$)'].quantile(0.25)
Q3 = df['GDP (current US$)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[~((df['GDP (current US$)'] < lower_bound) | (df['GDP (current US$)'] > upper_bound))]

df_cleaned = df
```

## Data wrangling

### Subtask:
Prepare the data for analysis by creating new features and transforming existing ones.

Analyze the average values and trends of key indicators for each country, calculate descriptive statistics, and investigate relationships between key indicators using scatter plots.

**Reasoning**:
Calculate the correlation matrix, visualize it as a heatmap, and analyze trends of key indicators over time for different countries.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation Analysis
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Identify highly correlated pairs (example: correlation > 0.8)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))
print("Highly Correlated Pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")


# Temporal Analysis (example: GDP growth over time for a few countries)
countries_to_plot = ['usa', 'chn', 'ind', 'jpn', 'deu']
gdp_growth_over_time = df[df['country'].isin(countries_to_plot)][['country', 'year', 'GDP growth (annual %)']]

plt.figure(figsize=(10, 6))
for country in countries_to_plot:
    country_data = gdp_growth_over_time[gdp_growth_over_time['country'] == country]
    plt.plot(country_data['year'], country_data['GDP growth (annual %)'], label=country)

plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.title('GDP Growth Over Time for Selected Countries')
plt.legend()
plt.grid(True)
plt.show()
```

**Reasoning**:
Create new features 'GDP_growth_rate' and 'income_group' based on existing columns, and save the updated DataFrame to a new CSV file.



```python
# Calculate GDP growth rate
df_cleaned['GDP_growth_rate'] = 0.0  # Initialize the new column

for country in df_cleaned['country'].unique():
    country_data = df_cleaned[df_cleaned['country'] == country]
    for i in range(1, len(country_data)):
        current_gdp = country_data.iloc[i]['GDP (current US$)']
        previous_gdp = country_data.iloc[i-1]['GDP (current US$)']
        if previous_gdp != 0:
            growth_rate = (current_gdp - previous_gdp) / previous_gdp * 100
            df_cleaned.loc[country_data.index[i], 'GDP_growth_rate'] = growth_rate


# Categorize countries into income groups based on GDP per capita
# Define income group thresholds (example values - adjust as needed)
low_income_threshold = 1045
middle_income_threshold = 4095

df_cleaned['income_group'] = 'Unknown'  # Initialize the new column

df_cleaned.loc[df_cleaned['GDP per capita (current US$)'] < low_income_threshold, 'income_group'] = 'Low income'
df_cleaned.loc[(df_cleaned['GDP per capita (current US$)'] >= low_income_threshold) & (df_cleaned['GDP per capita (current US$)'] < middle_income_threshold), 'income_group'] = 'Lower-middle income'
df_cleaned.loc[df_cleaned['GDP per capita (current US$)'] >= middle_income_threshold, 'income_group'] = 'Upper-middle income'

# Save the updated DataFrame to a new CSV file
df_cleaned.to_csv('world_bank_indicators_wrangled.csv', index=False)
```

## Data analysis

### Subtask:
Analyze the relationships between different World Bank indicators across countries and over time. Focus on the newly created features and explore potential correlations with other indicators.


**Reasoning**:
Analyze the relationships between GDP growth rate and other economic indicators, and visualize these relationships using scatter plots. Also, analyze how income group classification relates to other indicators and visualize using box plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Relationship between GDP growth rate and other economic indicators
indicators = ['GDP per capita (current US$)', 'Population, total']
plt.figure(figsize=(15, 5))
for i, indicator in enumerate(indicators):
    plt.subplot(1, 2, i + 1)
    sns.scatterplot(x=indicator, y='GDP_growth_rate', data=df_cleaned, hue='income_group')
    plt.title(f'GDP Growth Rate vs. {indicator}')
    plt.xlabel(indicator)
    plt.ylabel('GDP Growth Rate')
plt.tight_layout()
plt.show()

correlation_gdp = df_cleaned[['GDP_growth_rate', 'GDP per capita (current US$)', 'Population, total']].corr()
display(correlation_gdp)


# 2. Income group vs. other indicators
indicators = ['Access to electricity (% of population)', 'Life expectancy at birth, total (years)', 'School enrollment, primary (% gross)']
plt.figure(figsize=(15, 5))
for i, indicator in enumerate(indicators):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='income_group', y=indicator, data=df_cleaned)
    plt.title(f'{indicator} by Income Group')
    plt.xlabel('Income Group')
    plt.ylabel(indicator)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Reasoning**:
Investigate trends in key indicators over time for different income groups, and visualize these trends using line plots.



```python
# Group data by year and income group
grouped_data = df_cleaned.groupby(['year', 'income_group'])

# Calculate the average values of indicators
average_indicators = grouped_data[['GDP per capita (current US$)', 'Life expectancy at birth, total (years)', 'Access to electricity (% of population)']].mean().reset_index()

# Visualize trends using line plots
plt.figure(figsize=(15, 8))
for indicator in ['GDP per capita (current US$)', 'Life expectancy at birth, total (years)', 'Access to electricity (% of population)']:
    plt.plot(average_indicators[average_indicators['income_group'] == 'Low income']['year'], average_indicators[average_indicators['income_group'] == 'Low income'][indicator], label='Low income')
    plt.plot(average_indicators[average_indicators['income_group'] == 'Lower-middle income']['year'], average_indicators[average_indicators['income_group'] == 'Lower-middle income'][indicator], label='Lower-middle income')
    plt.plot(average_indicators[average_indicators['income_group'] == 'Upper-middle income']['year'], average_indicators[average_indicators['income_group'] == 'Upper-middle income'][indicator], label='Upper-middle income')
    plt.title(f'Trend of {indicator} over Time by Income Group')
    plt.xlabel('Year')
    plt.ylabel(indicator)
    plt.legend()
    plt.show()
```
